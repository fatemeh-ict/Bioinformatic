```python
def rotations(t):
  #Return list of rotation of input string t
  tt=t*2
  return [tt[i:i+len(t)] for i in range (0,len(t))]
```

```python
rotations('cat')
```

```python
def bwm(t):
  #return lexicographically sorted list of t's rotations
  return sorted(rotations(t))
```

```python
bwm('babaabdad$')
```

```python
print('\n'.join(bwm('babaabdad$')))
```

```python
def bwmviabwm(t):
  #given t, return bwm(t) by way of the bwm
  return "".join(map(lambda x:x[-1],bwm(t)))
```

```python
bwmviabwm('babaabdad$')
```
